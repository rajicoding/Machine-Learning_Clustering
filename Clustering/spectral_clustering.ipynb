```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X=dataset.iloc[:,[3,4]].values
X
```

```python
plt.scatter(X[:, 0], X[:, 1], s=30, c='gray', edgecolor='k')
plt.title("Original Data")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
Sc=StandardScaler()
X_processed=Sc.fit_transform(X)
X_processed
```

```python
# model for clustering with rbf as  affinity
from sklearn.cluster import SpectralClustering
spectral_model=SpectralClustering(n_clusters=4,affinity='rbf',assign_labels='discretize').fit(X_processed)
```

```python
spectral_model
```

```python
labels=spectral_model.fit_predict(X_processed)
```

```python
labels
```

```python
#Scatter Plot: Each cluster is plotted with a unique label using a loop over np.unique(labels).
#Legend: The plt.legend() function adds a legend to identify clusters.
#This approach ensures that each cluster is visually distinct and labeled appropriately in the legend.

for cluster_label in np.unique(labels):
    cluster_points = X[labels == cluster_label]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_label}')

plt.title('Spectral Clustering Results')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()  # Add legend
plt.show()
```

Spectral Clustering with affinity as poly

```python
# model for clustering with affinity as poly
from sklearn.cluster import SpectralClustering
spectral_model_1=SpectralClustering(n_clusters=4,affinity='poly',degree=3)
```

```python
labels_1=spectral_model_1.fit_predict(X_processed)
```

```python
labels_1
```

```python
#Scatter Plot: Each cluster is plotted with a unique label using a loop over np.unique(labels).
#Legend: The plt.legend() function adds a legend to identify clusters.
#This approach ensures that each cluster is visually distinct and labeled appropriately in the legend.

for cluster_label in np.unique(labels_1):
    cluster_points = X[labels_1 == cluster_label]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_label}')

plt.title('Spectral Clustering Results')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()  # Add legend
plt.show()



```
