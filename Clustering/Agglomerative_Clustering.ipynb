```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X= dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
# dendogram - no of clusters
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X, method='ward'))
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
# creating agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering
```

```python
# creating the model and clustering each data point in the dataset 
clusmodel=AgglomerativeClustering(n_clusters=5)
label=clusmodel.fit_predict(X)
```

```python
# dataset is loaded to supervised and the column cluster_group is added to the same
supervised=pd.DataFrame(dataset)
```

```python
supervised["cluster_group"]=label
```

```python
supervised
```

```python
# plotting the clusters for each datapoints
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],
                 fit_reg=False, legend=True, legend_out=True)
```
