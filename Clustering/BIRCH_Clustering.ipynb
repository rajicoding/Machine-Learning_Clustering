```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X=dataset.iloc[:,[3,4]].values
X
```

```python
# model creation for clustering
from sklearn.cluster import Birch
```

```python
birch_model = Birch(branching_factor=50,n_clusters=4,threshold=1.5).fit(X)
birch_model
```

```python
birch_model.labels_
```

```python
y=birch_model.predict(X)
```

```python
y
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised["clustered_group"] = y
```

```python
supervised
```

```python
#visualizing the clusters

plt.scatter(X[:,0],X[:,1], c = y, cmap = 'rainbow', alpha = 0.7, edgecolors = 'b')
plt.title("BIRCH Clustering")
plt.xlabel("Annual income")
plt.ylabel("Spending Score")
plt.show()
```

```python
labels=birch_model.labels_
plt.figure(figsize=(8, 6))
for cluster in set(labels):
    cluster_points = X[labels == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster}')

plt.title("Birch Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

```
