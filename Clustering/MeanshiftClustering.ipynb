```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X=dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
# clustering using meanshift
from sklearn.cluster import MeanShift,estimate_bandwidth

```

```python
# using estimate_bandwidth, ideal bandwidth value can  be calculated by  mentioning quantile and n_samples value
# quantile= 0.15 means 15 percent of distance
# n_samples= subset size used for bandwidth estimation
bandwidth = estimate_bandwidth(X, quantile=0.15, n_samples=200)
print("estimated_bandwidth",bandwidth)

ms = MeanShift(bandwidth=bandwidth)
ms.fit(X)
labels = ms.labels_
cluster_centers = ms.cluster_centers_

labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("number of estimated clusters : %d" % n_clusters_)
```

```python
y= ms.predict(X)
```

```python
y
```

```python
import matplotlib.pyplot as plt

plt.figure(1)
plt.clf()
unique_labels = np.unique(labels)

colors = ["#dede00", "#377eb8", "#f781bf",'#FF0000','#FF00FF']
markers = ["x", "o", "^","<","*"]

for k, col in zip(unique_labels, colors):
    my_members = labels == k
    cluster_center = cluster_centers[k]
    plt.plot(X[my_members, 0], X[my_members, 1], markers[k], color=col,label=f'Cluster {k}')
    plt.plot(
        cluster_center[0],
        cluster_center[1],
        markers[k],
        markerfacecolor=col,
        markeredgecolor="k",
        markersize=14,
    )
plt.title("Estimated number of clusters for Meanshift: %d" % n_clusters_)
plt.legend()
plt.show()
```
